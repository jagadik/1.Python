```python
# Catagory the people by their age like children, adult, citizen, senior citizen...
age = int(input("age:"))

if(age < 18):
    print("Age : ", age)
    print("Child")
elif(age < 35):
    print("Age : ", age)
    print("Adult")
elif(age < 50):
    print("Age : ", age)
    print("Citizen")
else:
    print("Age : ", age)
    print("Senior Citizen")
```

```python
# Check whether the given number is divisible by 5

number2 = int(input("Enter a number to check if it is divisible by 5:"))

if(number2%5 == 0):
    print("Divisible by 5")
else:
    print("Not divisible by 5")
```
