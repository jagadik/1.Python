```python
class OddEvenChecker():
    def secondFunc():
        number = int(input ("Enter the number:"))
        if (number%2 == 1):
            print (number,"is ODD number")
        else:
            print (number,"is Even number")
```

```python
OddEvenChecker.secondFunc()
```

```python
class ElegiblityForMarriage():

    def thridFunc(gender, age):

        print("AGE : ", age)
        print("GENDER :", gender)
        
        if (gender=="male" and age>21):
            print("Eligible for Marriage")
        elif (gender=="male" and age<21):
            print("NOT Eligible for Marriage")
        elif (gender=="female" and age>18):
            print("Eligible for Marriage")
        elif (gender=="female" and age<21):
            print("NOT Eligible for Marriage")
        else:
            print ("Invalid Input")
        
```

```python
ElegiblityForMarriage.thridFunc('male', 20)
```

```python
class FindPercent():
    
    def fourthFunc(s1, s2, s3, s4, s5):

        print("Subject 1 = ", s1)
        print("Subject 2 = ", s2)
        print("Subject 3 = ", s3)
        print("Subject 4 = ", s4)
        print("Subject 5 = ", s5)
        
        total = s1 + s2 + s3 + s4 + s5
        percent = total/5
        print ("Total : ",total)
        print ("Percentage : ",percent)
```

```python
FindPercent.fourthFunc(65, 89, 68, 78, 96)
```

```python
class Triangle():

    def fifthFunc():
        #Area of triangle
        height = int(input ("Height:"))
        breadth = int(input ("Breadth:"))

        print("Height: ", height)
        print("Breadth: ", breadth)
        print("Area formula: (Height*Breadth)/2")
        print("Area of Triangle: ", (height * breadth)/2)

        #Perimeter of triangle
        height1 = int(input ("Height1:"))
        height2 = int(input ("Height2:"))
        breadth1 = int(input ("Breadth1:"))
        
        perimeter= height1 + height2 + breadth1
        print("Perimeter formula: Height1 + Height2 + Breadth")
        print ("Perimeter of Triangle: ",perimeter)
        
```

```python
Triangle.fifthFunc()
```
