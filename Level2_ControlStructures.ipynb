```python
list = (20,10,16,19,25,1,276,188)

print(list)

print("Odd numbers:")
for i in list:
    if i % 2 == 1:
        print(i)
```

```python
print(list)

print("Even numbers:")
for j in list:
    if j % 2 == 0:
        print(j)
```
